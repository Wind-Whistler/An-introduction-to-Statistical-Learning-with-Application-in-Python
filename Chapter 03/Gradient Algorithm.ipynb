Use BGD To Solve Linear Regression

```python
import numpy as np

# initialize the data
np.random.seed(123)
x1 = np.random.normal(loc=5, scale=5, size=100)
x2 = np.random.normal(loc=10, scale=5, size=100)
x3 = np.random.normal(loc=20, scale=5, size=100)
x0 = np.ones(100)  # np.ones create an array like [1,1,1,1,1,1,1...,1]
x = np.array([x0, x1, x2, x3])  # X is 4*100 array
true_params = np.array([20, 3, 5, -7])
y = true_params @ x + np.random.normal(scale=5, size=100)  # Y=X * beta + epsilon

# initialize the para
sp = np.array([20, 3,5, 5]) 
learning_rate = 0.0001
n_samples = 100
y_hat = sp @ x
e = y - y_hat
error = np.mean(e**2) 

# Gradient Down
max_iterations = 1000000
iteration = 0

while error >= 0.001 and iteration < max_iterations:
    gradient = -2 * x @ e / n_samples
    sp = sp - learning_rate * gradient
    y_hat = sp @ x
    e = y - y_hat
    error = np.mean(e**2)
    iteration += 1
    if iteration % 1000 == 0:
        print(f"Iteration {iteration}, Error: {error:.6f}")
print("\nFinal Result:", sp)
print("True Result:", true_params)
print("SE:", error)
print("Iteration:", iteration)
```

Use Matrix to Solve Linear Regression

```python
# initialize the data
np.random.seed(123)
x1 = np.random.normal(loc=5, scale=5, size=100)
x2 = np.random.normal(loc=10, scale=5, size=100)
x3 = np.random.normal(loc=20, scale=5, size=100)
x0 = np.ones(100)  # np.ones create an array like [1,1,1,1,1,1,1...,1]
x = np.array([x0, x1, x2, x3])  # X is 4*100 array
true_params = np.array([20, 3, 5, -7])
y = true_params @ x + np.random.normal(scale=1, size=100)  # Y=X * beta + epsilon

P=np.linalg.inv(x @ x.T) @x@y
print(P)
```
