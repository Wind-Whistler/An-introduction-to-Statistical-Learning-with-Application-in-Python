```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("Boston.csv")
print(data.head())
data.describe()
```

```python
for i in range(0,data.shape[1]-1):
    data.iloc[:,i]=(data.iloc[:,i]-np.mean(data.iloc[:,i]))/np.std(data.iloc[:,i]);
data.describe()
```

```python
def Loss(data,beta):
    x=data[:,0:13]
    E = np.ones((x.shape[0],1))
    x=np.hstack([E, x]) 
    y=data[:,13]
    l=len(data)//5
    mbeta=x@beta
    y_hat=np.column_stack([mbeta[0:l,0],mbeta[l:2*l,1],mbeta[2*l:3*l,2],mbeta[3*l:4*l,3],mbeta[4*l:5*l,4]])
    y=np.column_stack([y[0:l],y[l:2*l],y[2*l:3*l],y[3*l:4*l],y[4*l:5*l]])
    m=0  
    for i in range(0,5):
        m=m+np.sum((y[:,i]-y_hat[:,i])**2)/l
    return m/5
    
#针对5-CV的特别梯度函数
def Gradient(data,beta):
    x=data[:,0:13]
    E = np.ones((x.shape[0],1))
    x=np.hstack([E, x]) 
    l=len(data)//5
    y=data[:,13].reshape(len(data),1)
    mbeta=x@beta
    y_hat=np.column_stack([mbeta[0:l,0],mbeta[l:2*l,1],mbeta[2*l:3*l,2],mbeta[3*l:4*l,3],mbeta[4*l:5*l,4]])
    y=np.column_stack([y[0:l,0],y[l:2*l,0],y[2*l:3*l,0],y[3*l:4*l,0],y[4*l:5*l,0]])  
    x=np.column_stack([x[0:l,:],x[l:2*l,:],x[2*l:3*l,:],x[3*l:4*l,:],x[4*l:5*l,:]])
    G=-2/l*x.T@(y-y_hat)
    Gradient=np.column_stack([G[0:14,0],G[14:28,1],G[28:42,2],G[42:56,3],G[56:70,4]])  #14*5的梯度矩阵
    return Gradient
def CV(data):  
    np.random.seed(123)
    Stochastic_Data = data.sample(frac=1, random_state=123).reset_index(drop=True)
    l=len(Stochastic_Data)//5
    fold_1= np.array(Stochastic_Data.iloc[0:l,:])
    fold_2= np.array(Stochastic_Data.iloc[l:2*l,:])
    fold_3= np.array(Stochastic_Data.iloc[2*l:3*l,:])
    fold_4= np.array(Stochastic_Data.iloc[3*l:4*l,:])
    fold_5= np.array(Stochastic_Data.iloc[4*l:5*l,:])
    M1_Tr=np.vstack([fold_1,fold_2,fold_3,fold_4])
    M2_Tr=np.vstack([fold_2,fold_3,fold_4,fold_5])
    M3_Tr=np.vstack([fold_3,fold_4,fold_5,fold_1])
    M4_Tr=np.vstack([fold_4,fold_5,fold_1,fold_2])
    M5_Tr=np.vstack([fold_5,fold_1,fold_2,fold_3])
    #训练矩阵
    M_Tr=np.vstack([M1_Tr,M2_Tr,M3_Tr,M4_Tr,M5_Tr])
    M1_Te=np.array(fold_5)
    M2_Te=np.array(fold_1)
    M3_Te=np.array(fold_2)
    M4_Te=np.array(fold_3)
    M5_Te=np.array(fold_4)
    #测试矩阵
    M_Te=np.vstack([M1_Te,M2_Te,M3_Te,M4_Te,M5_Te])
    M=[M_Tr,M_Te]
    return M
def D(beta):
    Des=np.zeros(5)
    for i in range(0,beta.shape[1]):
        Des[i]=np.sqrt(np.sum(beta[:,i]**2)) 
    return max(Des)  
```

```python
M_tr=CV(data)[0]
M_te=CV(data)[1]
l=[10**(-3+0.1*i) for i in range(0,60)]
MSE=[0]*60
miu=0.1
for i in range(0,60):
    beta_p1 = np.zeros((14, 5))      
    beta_p2 = np.full((14, 5), 100.0)  
    iteration=0
    while D(beta_p1-beta_p2)>0.0001 and iteration<10000:
        beta_p1=beta_p2.copy()
        G=Gradient(M_tr,beta_p1)
        beta_mid=beta_p1-miu*G
        index_p = beta_mid > l[i]*miu
        beta_p2[index_p] = beta_mid[index_p] -l[i]*miu
        index_n = beta_mid < -l[i]*miu
        beta_p2[index_n] = beta_mid[index_n] +l[i]*miu
        index_z=np.abs(beta_mid)<=l[i]*miu
        beta_p2[index_z]=0
        iteration=iteration+1
        if iteration%100==0:
            print(f'i={i}')
            print(f'iteration:{iteration}')
            print(D(beta_p1-beta_p2))
    MSE[i]=Loss(M_te,beta_p1)   
print(MSE)
index=MSE.index(min(MSE))
figure=plt.figure(figsize=(8, 6))
plt.semilogx(l,MSE,color="blue",label="Lambda vs MSE")
plt.axvline(x=l[index], color='green', linestyle='--',label='Optimal Step')
plt.xlabel("Lambda")
plt.ylabel("MSE")
plt.legend()
plt.savefig("CV.png")
plt.show()
print(l[index])
```
